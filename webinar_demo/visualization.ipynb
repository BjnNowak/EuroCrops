```python
shpfile = geopandas.read_file(shpPath+shpName+'.shp')
```

```python
shpfile.plot(column='EC_hcat_n')
plt.show()
```

```python
from collections import Counter

crops = Counter(shpfile['EC_hcat_n']).keys()
count = Counter(shpfile['EC_hcat_n']).values()
```

```python
fig, ax = plt.subplots()
fig.set_figheight(21)
ax.barh(range(1,len(count)+1), list(count))
ax.set_yticks(range(1,len(count)+1),labels = list(crops))
plt.show()
```
